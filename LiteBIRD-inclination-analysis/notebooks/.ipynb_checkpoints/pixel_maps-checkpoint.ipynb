```python
try:
    import numpy as np
    import toml
    from matplotlib import pyplot as plt
    from os import listdir
    from os.path import isfile, join
    from statistics import mean
    print("Imports completed")
except Exception as E:
    print("Unable to import a package")
    print(E)

try:
    base_dir = "../results/maps/"
    configfile = "../parameters.toml"
    config_params = toml.load(configfile)
    print("Macros defined.")
    print("Parameters: ",config_params)
except Exception as E:
    print("Unable to define a macro")
    print(E)

data = []
```

```python
for p in config_params["simulation"]["planets"]:
    path = base_dir+p+"/"
    files = [f for f in listdir(path) if isfile(join(path, f))]
    for f in files:
        print("Reading file: ",f)
        error = list(np.loadtxt(path+f,usecols=(0)))
        data.append((p,error))
        print("Data appended!\n")
```

```python
fig, ax = plt.subplots(1,3,figsize=(14,6))
for i,d in enumerate(data):
    ax[i].set_title(d[0].capitalize())
    ax[i].set_xlabel("Error")
    ax[i].set_ylabel("Frequency")
    ax[i].axvline(mean(d[1]), color='orange', linestyle='dashed', linewidth=1)
    ax[i].hist(d[1],bins=100)
```
